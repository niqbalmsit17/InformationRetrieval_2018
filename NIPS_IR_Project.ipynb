```python
from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(df[df.columns[1:257]], test_size=0.20, random_state=0)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=1, n_init=50).fit(X_train)
```

```python
predictions = kmeans.predict(X_test)
```
