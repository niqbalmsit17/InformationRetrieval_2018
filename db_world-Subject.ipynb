```python
import pandas as pd
import numpy as np
import csv
import sys
import os

df = pd.read_csv('dbworld_subjects.csv')
```

```python
from sklearn.model_selection import train_test_split

#n_array = df.values
#print(n_array.shape)
#n_array.reshape(243, 64)
#df.CLASS

#df[df.columns[0:242]]
#print(y.size)
X_train, X_test, y_train, y_test = train_test_split(df[df.columns[0:242]], df.CLASS, test_size=0.20, random_state=0)
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print ("Accuracy Rate, which is calculated by accuracy_score() is: %f" % accuracy_score(y_test, y_pred))
```

# Rocchio Algorithm

```python
from sklearn.neighbors.nearest_centroid import NearestCentroid


model = NearestCentroid(metric='euclidean')

model.fit(X_train, y_train)
 
# evaluate the model and update the accuracies list
score = model.score(X_test, y_test)
print("accuracy=%.2f%%" % (score * 100))
```

```python
predictions = model.predict(X_test)
 
# show a final classification report demonstrating the accuracy of the classifier
# for each of the digits
print("EVALUATION ON TESTING DATA")
print(classification_report(y_test, predictions))
```

```python
def plot_confusion_matrix(cm,class_,title='Confusion matrix',cmap=plt.cm.Blues):
    """
    This function plots a confusion matrix
    """
    fig, ax = plt.subplots(figsize=(5, 5))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    tick_marks = np.arange(len(class_))
    plt.xticks(tick_marks, class_, rotation=90)
    plt.yticks(tick_marks, class_)

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True labels')
    plt.xlabel('Predicted labels')
    plt.show()
```

```python
plot_confusion_matrix(confusion_matrix(y_test, predictions), ["0", "1"])
```

# K Nearest Neighbor Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

for k in range(1, 16, 2):
    model = KNeighborsClassifier(n_neighbors=k, weights='distance')

    model.fit(X_train, y_train)
 
    # evaluate the model and update the accuracies list
    score = model.score(X_test, y_test)
    print("k=%d, accuracy=%.2f%%" % (k, score * 100))
```

```python
from sklearn.metrics import classification_report

# re-train our classifier using the best k value and predict the labels of the
# test data
model = KNeighborsClassifier(n_neighbors=9)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
 
# show a final classification report demonstrating the accuracy of the classifier
# for each of the digits
print("EVALUATION ON TESTING DATA")
print(classification_report(y_test, predictions))
```

```python
plot_confusion_matrix(confusion_matrix(y_test, predictions), ["0", "1"])
```
